```
from sklearn.datasets import load_iris
iris = load_iris()
iris
```


![Image of iris](https://evolution.berkeley.edu/evolibrary/images/interviews/flower_diagram.gif)

![Image of iris](https://cdn-images-1.medium.com/max/2000/1*7bnLKsChXq94QjtAiRn40w.png)

```
from sklearn.tree import DecisionTreeClassifier, export_graphviz
import pydotplus
from IPython.display import Image

clf = DecisionTreeClassifier().fit(iris.data, iris.target)
dot_data = export_graphviz(clf, out_file=None, filled=True, rounded=True,
                                feature_names=iris.feature_names,  
                                class_names=['Versicolor','Setosa','Virginica'])
graph = pydotplus.graph_from_dot_data(dot_data)  
image=graph.create_png()
graph.write_png("kmc_dt.png")
Image(filename="kmc_dt.png")
```
