```
import tweepy
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
```

```
#My Twitter API Authentication Variables
consumer_key = 'Get your own key'
consumer_secret = 'Get your own key'
access_token = 'Get your own key'
access_token_secret = 'Get your own key'

```

```
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth)

tweets = api.search('#Tseriesvspewdiepie', count=100)


data = pd.DataFrame(data=[tweet.text for tweet in tweets], columns=['Tweets'])

display(data.head(10))


print(tweets[0].created_at)
```

```
import nltk
nltk.download('vader_lexicon')
```

```
sid = SentimentIntensityAnalyzer()


listy = []

for index, row in data.iterrows():
  ss = sid.polarity_scores(row["Tweets"])
  listy.append(ss)
  
se = pd.Series(listy)
data['polarity'] = se.values

display(data.head(100))
```

```
neg, neu, pos = 0, 0, 0
```

```
for i in data['polarity']:
  neg += i['neg']
  neu += i['neu']
  pos += i['pos']
```

```
labels = 'negative', 'neutral', 'positive'
sizes = [neg, neu, pos]
explode = (0.1, 0, 0)
colors = ['#ef4b40', '#d4bff9', '#75e900']
```

```
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=45, colors=colors)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title(f"Sentiment Analysis for the Pewdiepie vs T-series tweets ")
fig1 = plt.gcf()

plt.savefig(f'result.png', dpi=250)
plt.show()


```
