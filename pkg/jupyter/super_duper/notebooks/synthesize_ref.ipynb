```
# Load pretrained models
from notebook_utils.synthesize import Synthesizer
import IPython.display as ipd
synthesizer = Synthesizer(tts_path='forward_step80k.pt', 
                          voc_path='wave_step575k.pt')
```

##Synthesize

```
# Synthesize with melgan (alpha=1.0)
input_text = 'There is a way to measure the acute emotional intelligence that has never gone out of style.'
wav = synthesizer(input_text, voc_model='melgan', alpha=1)
ipd.Audio(wav, rate=synthesizer.dsp.sample_rate)
```

```
# Synthesize with wavernn (alpha=1.0)
input_text = 'There is a way to measure the acute emotional intelligence that has never gone out of style.'
wav = synthesizer(input_text, voc_model='wavernn', alpha=1)
ipd.Audio(wav, rate=synthesizer.dsp.sample_rate)

```

```
# Synthesize faster (alpha=1.2)
input_text = 'There is a way to measure the acute emotional intelligence that has never gone out of style.'
wav = synthesizer(input_text, voc_model='melgan', alpha=1.2)
ipd.Audio(wav, rate=synthesizer.dsp.sample_rate)
```

```
# Synthesize with amplified pitch
input_text = 'There is a way to measure the acute emotional intelligence that has never gone out of style.'
pitch_func = lambda x: x * 2.
wav = synthesizer(input_text, voc_model='melgan', alpha=1, pitch_function=pitch_func)
ipd.Audio(wav, rate=synthesizer.dsp.sample_rate)
```
