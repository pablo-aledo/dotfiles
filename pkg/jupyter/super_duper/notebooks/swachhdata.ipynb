## Text Module

```
import pandas as pd
from swachhdata.text import *
```

```
democrat = pd.read_csv('https://raw.githubusercontent.com/kritikseth/sentiment-analysis/master/twitter_democrat.csv')
republican = pd.read_csv('https://raw.githubusercontent.com/kritikseth/sentiment-analysis/master/twitter_republican.csv')
```

```
tweets = democrat['Tweets'].tolist() + republican['Tweets'].tolist()
```

```
tweets += tweets
```

```
len(tweets)
```

```
rec = urlRecast(process='remove')
rec.setup_recast(tweets)
```

```
rec.get_regex_
```

```
rec = htmlRecast()
rec.setup_recast(tweets[0])
```

```
rec = EscapeSequenceRecast()
rec.setup_recast(tweets[0])
```

```
rec = MentionRecast(process='extract_remove')
rec.setup_recast(tweets[0])
```

```
rec = ContractionsRecast()
rec.setup_recast(tweets[0])
```

```
rec = CaseRecast(process='fupper')
rec.setup_recast(tweets[0])
```

```
rec = EmojiRecast(process='extract_replace', space_out=True)
rec.setup_recast(tweets[778])
```

```
rec = HashtagRecast(process='extract_remove')
rec.setup_recast(tweets[6])
```

```
rec = ShortWordsRecast(min_length=3)
rec.setup_recast(tweets[0])
```

```
rec = StopWordsRecast(package='nltk')
rec.setup_recast(tweets[0])
```

```
rec = NumberRecast(process='extract_replace')
rec.setup_recast(tweets[17])
```

```
rec = AlphabetRecast()
rec.setup_recast(tweets[778])
```

```
rec = PunctuationRecast()
rec.setup_recast(tweets[409])
```

```
rec = TokenisationRecast(package='nltk', method='word')
print(rec.setup_recast(tweets[0]))
```

```
rec = StemmingRecast(package='ntlk', method='porter')
rec.setup_recast(tweets[0])
```

```
rec = LemmatizationRecast(package='nltk')
rec.setup_recast(tweets[0])
```
