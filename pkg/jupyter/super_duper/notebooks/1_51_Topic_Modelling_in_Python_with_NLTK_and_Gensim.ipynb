##### Clean Text and return Tokens

```python
#pip install spacy
#python -m spacy download en_core_web_sm
#copy en_core_web_sm from C:\Users\rajiv_kale\AppData\Local\Continuum\Anaconda3\Lib\site-packages
#to C:\Users\rajiv_kale\AppData\Local\Continuum\Anaconda3\Anaconda3\Lib\site-packages\spacy\data\

#=============
# On AWS EC2 we did this.
# Login as ec2-user
#pip install spacy
# python -m spacy download en_core_web_sm
# cd /home/ec2-user/anaconda3/envs/tensorflow_p36/lib/python3.6/site-packages/
# cp -R en_core_web_sm   spacy/data/

import spacy
spacy.load('en_core_web_sm')  # needs some installation and tweaking
from spacy.lang.en import English
parser = English()
def tokenize(text):
    lda_tokens = []
    tokens = parser(text)
    for token in tokens:
        if token.orth_.isspace():
            continue
        elif token.like_url:
            lda_tokens.append('URL')
        elif token.orth_.startswith('@'):
            lda_tokens.append('SCREEN_NAME')
        else:
            lda_tokens.append(token.lower_)
    return lda_tokens
```

##### wordnet and lemmatization

Use NLTK's wordnet to find meanings of words, synonyms, antonyms, and more. In addition, we use WordNetLemmatizer to get the root word.

```python
import nltk
#nltk.download('wordnet')
from nltk.corpus import wordnet as wn
def get_lemma(word):
    lemma = wn.morphy(word)
    if lemma is None:
        return word
    else:
        return lemma
    
from nltk.stem.wordnet import WordNetLemmatizer
def get_lemma2(word):
    return WordNetLemmatizer().lemmatize(word)
```

##### Filter out stop words

```python
#nltk.download('stopwords')
en_stop = set(nltk.corpus.stopwords.words('english'))
```

##### Define a function to prepare Text for Topic Modelling

```python
def prepare_text_for_lda(text):
    tokens = tokenize(text)
    tokens = [token for token in tokens if len(token) > 4]
    tokens = [token for token in tokens if token not in en_stop]
    tokens = [get_lemma(token) for token in tokens]
    return tokens
```

##### Look at the data that we have to work with

Open up our data, read line by line and print first 10 lines and print line count

```python
num_lines=0
with open('Datasets/dataset.csv') as f:
    for line in f:
        num_lines=num_lines+1
        if num_lines<10:
            print(line)
print("total number of lines=", num_lines)        
```

for each line, prepare text for LDA, then add to a list. Now we can see how our text data gets converted:

```python
#random.random(): Returns the next random floating point number in the range [0.0, 1.0].
#We choose subset of the data
import random
text_data = []
with open('Datasets/dataset.csv') as f:
    for line in f:
        tokens = prepare_text_for_lda(line)
        if random.random() > .79:
            print(tokens)
            text_data.append(tokens)
```

```python
len(text_data)
```

##### LDA with gensim

First, we are creating a dictionary from the data, then convert to bag-of-words corpus and save the dictionary and corpus for future use.

```python
from gensim import corpora
dictionary = corpora.Dictionary(text_data)
corpus = [dictionary.doc2bow(text) for text in text_data]
```

```python
# save corpus and dictionary to disk so that we can use it later during visualization
import pickle
pickle.dump(corpus, open('corpus.pkl', 'wb'))
dictionary.save('dictionary.gensim')
```

##### Ask LDA to find 5 topics from the given data. Takes about 1 minute.

```python
import gensim
NUM_TOPICS = 5
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes=15)
ldamodel.save('models/model5.gensim')
```

###### Focus on Topics and Let us see top 4 words in each topic

```python
topics = ldamodel.print_topics(num_words=6)
for topic in topics:
    print(topic)
```

With LDA, we can see how words contribute to make each topic

##### Visualization using pyLDAvis

pyLDAvis is designed to help users interpret the topics in a topic model that has been fit to a corpus of text data.

The package extracts information from a fitted LDA topic model to inform an interactive web-based visualization.

```python
dictionary = gensim.corpora.Dictionary.load('dictionary.gensim')
corpus = pickle.load(open('corpus.pkl', 'rb'))
lda = gensim.models.ldamodel.LdaModel.load('models/model5.gensim')
```

```python
#pip install pyLDAvis==2.1.1
import pyLDAvis.gensim
lda_display = pyLDAvis.gensim.prepare(lda, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display)
```

Saliency: a measure of how much the term tells you about the topic.

Relevance: a weighted average of the probability of the word given the topic and the word given the topic normalized by the probability of the topic.

The size of the bubble measures the importance of the topics, relative to the data.

First, we got the most salient terms, means terms mostly tell us about what’s going on relative to the topics. We can also look at individual topic.

Let us visulaize for 10 topics model

```python
lda10 = gensim.models.ldamodel.LdaModel.load('models/model10.gensim')
lda_display10 = pyLDAvis.gensim.prepare(lda10, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_display10)
```

Each circle represents a topic and Area of each circle gives topic prevalenve. Number 1 circle is most popular and last number 10 is least popular. Distance between each circle represents topic similarity.  On the right hand side are top 30 terms for selected topic. Blue bar shows overall term frequency and Red bar shows term frequency for a chosen topic.

Role of Lyamda: Changing Lyamda changes the terms used to represent topic. A setting for zero Lyamda means we chose Jargons (words execlusively used for this topic). Blue bars diminish.

See youtube videos such as https://www.youtube.com/watch?v=SF50IK5XgKA and https://www.youtube.com/watch?v=IksL96ls4o0 for details on interpreting pyLDAvis visulaization.

When we have 10 topics, we can see certain topics are clustered together more than when we had 5 topics,

this indicates the similarity between topics.
